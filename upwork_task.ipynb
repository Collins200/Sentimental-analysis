```python
# Remove all columns after column index 3
data.drop(data.iloc[:, 3:], inplace=True, axis=1)
```

```python
sep = '|'
punctuation_chars = '!"#%&\'()*+,-./:;<=>?@[\\]^_`{}~'
```

```python
mapping_table = str.maketrans(dict.fromkeys(punctuation_chars, ''))
```

```python
data['A'] = sep \
    .join(data['Question'].tolist()) \
    .translate(mapping_table) \
    .split(sep)
```

```python
data.head(10)
```

```python
import nltk

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]

data['C'] = data.A.apply(lemmatize_text)
```

```python
data['C']
```

```python
from nltk import pos_tag

data['C'] = data['C'].apply(pos_tag)
data
```
